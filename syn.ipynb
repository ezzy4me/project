```python
import librosa
import torch
import IPython.display as display
from transformers import Wav2Vec2ForCTC, Wav2Vec2Tokenizer
import numpy as np
```

```python
import os
os.chdir(r'/content')
os.getcwd()
```

```python
#load pre-trained model and tokenizer
tokenizer = Wav2Vec2Tokenizer.from_pretrained("facebook/wav2vec2-base-960h")
model = Wav2Vec2ForCTC.from_pretrained("facebook/wav2vec2-base-960h")
```

```python
#load audio file 
audio, sampling_rate = librosa.load("/content/MS-SNSD/NoisySpeech_training/noisy13_SNRdb_30.0_clnsp13.wav",sr=16000)
```

```python
# audio
display.Audio("/content/MS-SNSD/NoisySpeech_training/noisy13_SNRdb_30.0_clnsp13.wav", autoplay=True)
```

```python
input_values = tokenizer(audio, return_tensors = 'pt').input_values
input_values
```

```python
# store logits (non-normalized predictions)
logits = model(input_values).logits
logits
```

```python
# store predicted id's
# pass the logit values to softmax to get the predicted values
predicted_ids = torch.argmax(logits, dim =-1)
```

```python
# pass the prediction to the tokenzer decode to get the transcription
transcriptions = tokenizer.decode(predicted_ids[0])
```
